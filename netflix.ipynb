```python
from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt

# 1. Cargar el archivo netflix_titles.csv
data = pd.read_csv('/content/netflix_titles.csv')


# 2. Mostrar los primeros 10 registros
print('Muestra los primeros 10 registros')
print(data.head(10))
print()
print()

# 3. Mostrar los ultimos 15 registros
print('Muestra los ultimos 15 registros')
print(data.tail(20))
print()
print()
print()

# 4. Estadisticos Basicos
print("Mostrando estadistico basicos")
print(data.describe())
print()
print()
print()



# 5. Completar todos los datos vacíos (na) con ceros (0)
print("Mostrando si faltan balores")
print(data.isna())


# 6. Generar un gráfico de tipo barras que compare películas vs series desde el 2010 hasta el 2021.
# Filtrar los datos por año
# Calcular la cantidad de películas y series de TV lanzadas cada año

# Filtrar los datos para el período de 2010 a 2021
data = data[(data["release_year"] >= 2010) & (data["release_year"] <= 2021)]

# Agrupar los datos por año y tipo de contenido y contar la cantidad
data_por_año = data.groupby(["release_year", "type"]).size().unstack(fill_value=0)

# Crear un gráfico de barras múltiples
fig, ax = plt.subplots()

# Definir las posiciones de las barras para cada año
posiciones = range(len(data_por_año.index))

# Ancho de las barras
ancho_barras = 0.35

# Barras para las películas
ax.bar([pos - ancho_barras/2 for pos in posiciones], data_por_año["Movie"], width=ancho_barras, color="blue", label="Movie")

# Barras para las series de TV
ax.bar([pos + ancho_barras/2 for pos in posiciones], data_por_año["TV Show"], width=ancho_barras, color="orange", label="TV Show")

# Agregar etiquetas al eje X y al eje Y
ax.set_xlabel("Año")
ax.set_ylabel("Cantidad")

# Establecer las etiquetas del eje X como los años
ax.set_xticks(posiciones)
ax.set_xticklabels(data_por_año.index)

# Agregar leyenda
ax.legend(loc="upper right")

# Agregar un título al gráfico
plt.title("Comparación de películas vs. series de TV (2010-2021)")

# Mostrar el gráfico
plt.show()





```
